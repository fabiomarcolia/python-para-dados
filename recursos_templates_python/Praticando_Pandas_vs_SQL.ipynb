# Praticar e se familiarizar com Pandas e SQL
### Opera√ß√µes inspiradas em SQL para se familiarizar com Python
#### A ideia de usar PandasSQL √© para facilitar o uso do SQL diretamente no Notebook para aprender Pandas.

```python
import pandas as pd
from pandasql import sqldf

pysqldf = lambda q: sqldf(q, globals())
```

### Carregar os dados

```python
clientes = pd.read_excel('https://github.com/fabiomarcolia/python-para-dados/raw/refs/heads/main/dados/bases_treino/Clientes.xlsx')
enderecos = pd.read_excel('https://github.com/fabiomarcolia/python-para-dados/raw/refs/heads/main/dados/bases_treino/Enderecos.xlsx')
display(clientes)
display(enderecos)
```

### Tarefa 1: Selecionar as primeiras 10 linhas

```python
# Pandas
clientes.head(10)
```

```python
# PandasSQL
pysqldf('SELECT * FROM clientes LIMIT 10;')
```

### Tarefa 2: Filtrar resultados onde Idade > 18

```python
# Pandas
clientes[clientes['Idade'] > 18]
```

```python
# PandasSQL
pysqldf('SELECT * FROM clientes WHERE Idade > 18;')
```

### Tarefa 3: Ver estrutura/tipos de dados

```python
# Pandas
clientes.dtypes
```

```python
# PandasSQL
pysqldf('PRAGMA table_info(clientes);')
```

### Tarefa 4: Contar o n√∫mero de linhas

```python
# Pandas
clientes.shape[0]
```

```python
# PandasSQL
pysqldf('SELECT COUNT(*) FROM clientes;')
```

### Tarefa 5: Selecionar coluna 'Nome'

```python
# Pandas
clientes['Nome']
```

```python
# PandasSQL
pysqldf('SELECT Nome FROM clientes;')
```

### Tarefa 6: Classificar dados por coluna 'Nome'

```python
# Pandas
clientes.sort_values('Nome')
```

```python
# PandasSQL
pysqldf('SELECT * FROM clientes ORDER BY Nome;')
```

### Tarefa 7: Atualizar coluna 'Nome' para mai√∫sculas

```python
# Pandas
clientes['Nome'] = clientes['Nome'].str.upper()
clientes
```

```python
# PandasSQL
# Aten√ß√£o: APENAS EXEMPLO, IR√Å DAR ERRO EM RAZ√ÉO DE N√ÉO TER PERMISS√ÉO DE FAZER UPDATE NO GITHUB
pysqldf('UPDATE clientes SET Nome = UPPER(Nome);')
```

### Tarefa 8: Obter valores √∫nicos da coluna 'Nome'

```python
# Pandas
clientes['Nome'].unique()
```

```python
# PandasSQL
pysqldf('SELECT DISTINCT Nome FROM clientes;')
```

### Tarefa 9: Agrupar por 'Nome' e calcular soma de 'Total'

```python
# Pandas
clientes.groupby('Nome')['Total'].sum()
```

```python
# PandasSQL
pysqldf('SELECT Nome, SUM(Total) FROM clientes GROUP BY Nome;')
```

### Tarefa 10: Deletar coluna 'Endereco'

```python
# Pandas
clientes.drop(columns=['Endereco'], inplace=True)
clientes
```

```python
# PandasSQL
# Anten√ß√£o: Apenas exemplo. Sem permiss√£o de alterar o arquivo no Github.
pysqldf('ALTER TABLE clientes DROP COLUMN Endereco;')
```

### Tarefa 11: Renomear coluna 'Nome' para 'PrimeiroNome'

```python
# Pandas
clientes.rename(columns={'Nome': 'PrimeiroNome'}, inplace=True)
clientes
```

```python
# PandasSQL
pysqldf('ALTER TABLE clientes RENAME COLUMN Nome TO PrimeiroNome;')
```

### Tarefa 12: Inner Join entre clientes e endere√ßos

```python
# Pandas
clientes.merge(enderecos, on='IdCliente')
```

```python
# PandasSQL
pysqldf('SELECT * FROM clientes C JOIN enderecos E ON C.IdCliente = E.IdCliente;')
```
