# 02 — VS Code + Jupyter (fluxo de trabalho)

Objetivo: rodar notebooks e scripts no VS Code com o mínimo de atrito.

Tempo: ~15–20 min

## Checklist

1- Instalar extensões recomendadas (`.vscode/extensions.json`)
2- Selecionar interpretador: `.venv`
3- Rodar um notebook e um script
4- Exportar um notebook (HTML/PDF opcional)

> Dica: mantenha o terminal aberto na raiz do repo.


## Comandos úteis

- Ativar venv (Windows):

```bash
.venv\Scripts\activate
```

- Ativar venv (Linux/Mac):

```bash
source .venv/bin/activate
```

- Rodar script:

```bash
python scripts/03_pandas/02_groupby_pivot_agg.py
```


## Exercício rápido
