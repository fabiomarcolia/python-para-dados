```python
# Setup: detectar raiz do repositório + paths padrão
from pathlib import Path
import pandas as pd
import numpy as np

def find_repo_root(start: Path | None = None) -> Path:
    p = (start or Path.cwd()).resolve()
    for _ in range(12):
        if (p / "requirements.txt").exists() and (p / "README.md").exists():
            return p
        p = p.parent
    return (start or Path.cwd()).resolve()

ROOT = find_repo_root()
DATA_SAMPLE = ROOT / "data" / "sample"
DATA_SOURCE = ROOT / "data" / "source" / "bases-dados-analytics-powerbi-ml"

# Pasta do projeto (onde salvar assets/outputs/reports)
PROJ = ROOT / "projects" / "00_setup"
ASSETS = PROJ / "assets"
OUTPUTS = PROJ / "outputs"
REPORTS = PROJ / "reports"
for d in (ASSETS, OUTPUTS, REPORTS):
    d.mkdir(parents=True, exist_ok=True)

# Dataset padrão (sempre disponível)
sales_path = DATA_SAMPLE / "sales.csv"
customers_path = DATA_SAMPLE / "customers.csv"
sales = pd.read_csv(sales_path, parse_dates=["date"])
customers = pd.read_csv(customers_path, parse_dates=["signup_date"])

# Dataset real (opcional): se você adicionou o submodule/clone em dados/source/
has_real = DATA_SOURCE.exists()

print("ROOT:", ROOT)
print("Dataset sample:", sales.shape, customers.shape)
print("Dataset real disponível?", has_real)
```

```python
# Checagens rápidas
display(sales.head())
display(customers.head())

# Um gráfico simples para validar renderização e salvar em assets/
import matplotlib.pyplot as plt

by_region = sales.groupby("region", as_index=False)["revenue"].sum().sort_values("revenue", ascending=False)
plt.figure()
plt.bar(by_region["region"], by_region["revenue"])
plt.title("Receita por região (sample)")
plt.xticks(rotation=30, ha="right")
out = ASSETS / "quick_plot.png"
plt.tight_layout()
plt.savefig(out, dpi=160)
print("Salvo:", out)
plt.close()

# Um resumo em Markdown
top_product = sales.groupby("product", as_index=False)["revenue"].sum().sort_values("revenue", ascending=False).head(5)
md = []
md.append("# Quick Summary (00_setup)\n")
md.append(f"- Linhas sales: {len(sales)}\n")
md.append(f"- Linhas customers: {len(customers)}\n")
md.append("## Top 5 produtos por receita\n")
md.append(top_product.to_markdown(index=False))
report = REPORTS / "quick_summary.md"
report.write_text("\n".join(md), encoding="utf-8")
print("Salvo:", report)
```
